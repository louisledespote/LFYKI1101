## Q3

```
import csv
import matplotlib.pyplot as plt

t = []
T1 = []
T2 = []
dT = []

with open("tmp117_log_labo_6mars.csv", newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        t.append(float(row["Elapsed_Time_s"]))
        T1.append(float(row["T1_C"]))
        T2.append(float(row["T2_C"]))
        dT.append(float(row["Delta_T_C"]))

plt.figure()
plt.plot(t, T1, label="T1")
plt.plot(t, T2, label="T2")
plt.legend()
plt.grid(True)
plt.show()
```

## Q4

```
plt.figure()
plt.plot(t, dT, label="Delta T = T1 - T2")
plt.axhline(0, linestyle="--")
plt.legend()
plt.grid(True)
plt.show()
```
